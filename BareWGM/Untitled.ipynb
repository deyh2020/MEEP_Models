```python
import numpy as np
import matplotlib.pyplot as plt

n=np.linspace(1,200)

a=pow((1+(1/n)),n)
expo=np.ones(len(n))*np.exp(1)

fig,ax = plt.subplots(figsize=(16,10))

ax.plot(n,a,marker='')
ax.plot(n,expo,linestyle='--')


ax.set_xlabel('n')
ax.set_ylabel('$a_n$')
ax.set_title('Limit of $a_n=\\left(1+\\frac{1}{n}\\right)^n$')

fig.canvas.draw()

labels = [item for item in ax.get_yticks()]
print (labels)

labels[-1] = np.exp(1)

labels2= [str(np.round(label,decimals=2)) for label in labels]

labels2[-1]='big ol e'

plt.yticks(labels,labels2)

ax.grid()
plt.show()
```
