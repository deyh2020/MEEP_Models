```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt
import MeepFunctions as MF

```

```python
sim,src,tran = MF.buildModel(coreN=1,cladN=1,TL=10,Depth=40,GAP=100)


src_fr = mp.FluxRegion(center=mp.Vector3(0,0,0), size=mp.Vector3(0,1,0))                            
src = sim.add_flux(0.5, 8e-3, 100, src_fr)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.figure(dpi=200)
sim.plot2D(plot_sources_flag=True,plot_monitors_flag=True,eps_parameters={'interpolation':'none'})
plt.show()
```

```python
t = (1e-6/3e8)
tFactor = 1e-15/t # converts femptoseconds into unitless MEEP 

Simtime = 1e4 #in femptoseconds

sim.use_output_directory('Data/test')

#sim.reset_meep()
sim.run(mp.at_beginning(mp.output_epsilon),
        until=(Simtime*tFactor))
```

```python
plt.figure(dpi=200)
sim.plot2D(fields=mp.Ez,plot_sources_flag=True,plot_monitors_flag=True)
plt.show()
```

```python
fig,axes = plt.subplots(1,1,figsize=(16,9))




wl = 1/np.array(mp.get_flux_freqs(tran))

Tran = np.array(mp.get_fluxes(tran))

Src = np.array(mp.get_fluxes(src))

axes.plot(wl,Tran,label='Overall Transmission')
axes.plot(wl,Src,label='Overall Transmission')



#fig_format(axes,'','Wavelength / um','Relative Power','')
```
